Które z następujących zadań wymagają w Twojej opinii inteligencji od człowieka?

Ponieważ inteligencja rozumiana jest na wiele różnych sposobów, postaram się wykazać w jakim zakresie potrzebujemy jej w danym przypadku i czy moglibyśmy zastąpić ją sztuczną inteligencją.

*   wypełnianie deklaracji PIT -
Uważam, że wymaga to inteligencji od człowieka w minimalnym stonpniu. Natomiast potrzeba dużo uwagi, dokładności i znajomości procedur Urzędów Skarbowych. Użycie SI byłoby ryzykowne. Lepiej polegać w tym przypadku na ludzkim myśleniu i ewentualnie wspomagać się sztuczną inteligencją.
W ubiegłym roku zrobiłam tak rozliczając zagraniczną deklarację podatkową, sztuczna inteligencja nie była w stanie sobie z tym poradzić, ale mi też wcale nie było łatwo.

*   streszczanie tekstu - Zdecydowanie, by przenieść myśli autora i zredagować je, człowiek musi wykazać się inteligencją(i związaną z nią kreatywnością), by utrzymać sens logiczny pierwotnego tekstu.
*   układanie rozkładu jazdy transportu miejskiego - Nie jest to proste zadanie i wymaga dużej inteligencji, trzeba w tym przypadku myśleć do przodu, planować i przede wszystkim brać pod uwagę wiele zmiennych. Tutaj by wspomóc człowieka sztuczna inteligencja mogłaby być bardzo przydatna.

*   pisanie programów komputerowych - wymaga to sporej inteligencji u człowieka, myślenia logicznego i analitycznego, by nauczyć się pisania w języku programowania.
*   testowanie oprogramowania, by było efektywne wymaga inteligentnej osoby, która zauważy ewentualne błędy i będzie rozumiała program. Chyba, że chodzi tu o najprostsze programy, teoretycznie widzimy od razu czy program działa czy nie.

*   komponowanie muzyki - By stworzyć dobrą w mojej opinii muzykę, potrzebujemy inteligencji emocjonalej, której np. SI nie posiada. Niestety dziś jak widzimy po rynku muzycznym, inteligencja nie jest raczej wymagana.
*   wymienione operacje matematyczne na podstawowym stopniu wymagają niewiele inteligenji, ale jest ona potrzebna
*   prowadzenie samochodu - Dobrze by było gdyby ludzie używali inteligencji w postaci planowania i myślenia abstarkcyjnego. Samo prowadzenie to raczej kwestia wyuczenia się tej zdolności, ale by było to bezpieczne i by człowiek efektywnie mógł reagować w sytuacjach zagrożenia-potrzebuje inteligencji. Wiemy, że dziś ta funkcja bywa powierzana sztucznej inteligencji, ale uważam, że jest to jeden z ludzkich błędów by nasze życie zależało od SI.

**Zadanie 3**
Które z poniższych rodzajów komunikacyjnego zachowania człowieka mogą być
obecnie skutecznie imitowane przez sztuczne systemy (odpowiednio oprogramowane
maszyny):

* Rozmowa towarzyska - może i obecnie jest skutecznie imitowana jeśli chodzi o niezobowiązujące emocjonalnie tematy. AI nie rozumie kontekstów społecznych i kulturowych, mimo to były już sytuacje, w których rozmawiało z kimś w roli terapeuty. Myślę, że jest to związane z brakiem wydawania osądów przez AI i rozwijaniu zawsze tej myśli, na której zależy człowiekowi.
* Dyskusja polityczna i naukowa - tutaj zdecydowanie AI sobie nie radzi ze względu na wymieniony wyżej argument nierozumienia kontekstów kulturowo-społecznych. Nie rozumieją głębszego kontekstu. Brak sztucznym systemom kreatywnego i krytycznego postrzegania. Przede wszystkim nie nadają sie do dyskusji, bo nie są w stanie obrać stałego stanowiska. Oczywiście mogą być wsparciem, ale nie będą imitowały skutecznie takcih dyskusji.
* Odpowiadanie na pytania klientów telefonicznej infolinii - Przy mniej złożonych pytaniach AI sobie radzi, ale nie imituje skutecznie konsultanta.
* Odpowiadanie na pytania klientów internetowej infolinii - Tutaj sztuczne systemy sobie radzą, bo są od nich wymagane bardziej rutynowa zadania jak np.sprawdzanie statusu zamówienia.



**Zadanie 4**

1.   Spróbuj zdefiniować różnice pomiędzy botem udającym człowieka (przygotowywanym na test Turinga) a botem
„asystentem, służącym”.

Celem bota udającego człowieka jest imitowanie ludzkich zachowań, tak by oszukać rozmówcę, by ten uwierzył, że rozmawia z człowiekiem. Bot imitujący człowieka koncentruje się na naśladowaniu mowy, stylu wypowiedzi i emocji. Ma  brzmieć naturalnie i odpowiadać na szeroki zakres pytań zarówno technicznych jak i takich podobnych do rozmów towarzyskich.

Natomiast celem bota asystenta służącego jest wykonywanie określonych zadań, jest ograniczony w jakimś konkrenym zakrasie w zależności od jego roli.
Nie zawsze ma głęboko funkcjonalne umiejętności rozwiązywania problemów
Bot asystent koncentruje się na funkcjonalności, przydatności i skuteczności, a nie na imitacji ludzkich zachowań.
Nie udaje człowieka – może nawet wyraźnie identyfikować się jako bot – i nie stara się angażować w naturalne rozmowy.


2.   Sprawdź dwa boty z obu z tych rodzajów na występowanie zachowań i sporządź raport ze spostrzeżeń.

Jeśli chodzi o opowiadanie żartów to boty udające człowieka poradziły sobie bardzo dobrze(np. personalityforge i cleverbot), natomiast boty asystenci jak chatGPT nie rozumiały o co mi chodzi lub pisały żarty bez kontekstu.

Przytaczanie cytatów najlepiej wychodzi niektórym botom asystentom. Boty udające człowieka też są w stanie przytaczać cytaty, ale czasem je modyfikują.

Nawiązywanie wypowiedzi do słow kluczowych to mocna strona botów asystujących, nawiązują do tych słow i udzielają potrzebnych odpowedzi, jak np. wyszukanie konkretnego rodzaju mebli, o które zapytałam. Chaty naśladujące człowieka też sobie z tym radziły, ale znacznie słabiej, bo wyszukiwały konkretnych fraz, ale czasem nie dostosowywały ich do kontekstu rozmowy.

Z zadawaniem dużej liczby pytań lepiej poradziły sobie boty asystujące. Naprowadzały na kontekst swoimi własnymi pytaniami. Te naśladujące człowieka, zaczynały w pewnym momencie tracić kontekst rozmowy przez zbyt dużo informacji.

Zauważyłam, że boty asystenci same celowo powracają do początku rozmowy i sekwencyjnie powtarzają poprzednie frazy. Są zaprogramowane w konkretnym celu i prowadzą rozmowę tak, by trzymać się blisko tematu. Boty naśladujące zaczynały czasem kolejne wątki, niezwiązane z poprzednimi sugerując się użytym przeze mnie słowem, po czym gubiły się w konktekście i dopiero wtedy powtarzały pytanie.

Boty asystenckie prawie cały czas zadawały pytania wynikające z moich odpowiedzi w logiczny sposób np. chciałam kupić nowe łóżko więc bot co chwile pytał "jaki kolor łóżka Cię interesuje? jaki rozmiar łóżka Cię interesuje?
Boty naśladujące zadawały pytania, które miały podtrzymać rozmowę, ale niekoniecznie nawiązywały do mojego pytania po dłuższej rozmowie.

Odpowiedzi ogólnikowe i wymijające to raczej dziedzina botów naśladujących, czasem brakuje im wiedzy w konkretnych tematach i odpisują 'trudno powiedzieć' lub 'to zależy od róznych czynników. Boty asystujące odpowiadają raczej konkretnie, ale w zakresie tematów do których zostały stworzone.

Boty asystujące nie zmieniają tematu rozmowy, robiły to dopiero gdy ja zadawałam pytanie na inny temat. Naśladujące zmieniały wątki samoistnie.
